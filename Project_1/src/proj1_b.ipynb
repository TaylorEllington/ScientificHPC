```python
import csv

with open('n.txt', 'rb') as file:
    n = file.read().replace('\n', '').split('  ')
    n.remove('')
    file.close()
    
with open('h.txt', 'rb') as file:
    h = file.read().replace('\n', '').split('  ')
    h.remove('')
    file.close()
    
with open('r.txt', 'rb') as file:
    r = file.read().replace('\n', '').split('  ')
    r.remove('')
    file.close()
    
with open('R.txt', 'rb') as file:
    R = file.read().replace('\n', '').split('  ')
    R.remove('')
    file.close()


    
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np




fig = plt.figure()
ax = fig.add_subplot(1,1,1)


plt.xlabel( 'n'  )

line, = ax.semilogy(n, r, color='blue', label="Relative Error", lw=2)
line2, = ax.semilogy(n ,R, color='red', lw=2, label="Upper Bound on Relative Error")


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


plt.show()
```
