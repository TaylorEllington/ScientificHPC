```python
import numpy as np
t = np.loadtxt("t.txt")
y = np.loadtxt("y.txt")
x = np.loadtxt("x.txt")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(x, t)
plt.xlabel("x")
plt.ylabel("t")

plt.show()
```
